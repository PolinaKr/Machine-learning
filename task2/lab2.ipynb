### Обработка данных

```python
import numpy as np
from sklearn import *
data=data[[ 'VehicleAge','IsBadBuy','Transmission','VehOdo','MMRAcquisitionAuctionAveragePrice',
           'MMRAcquisitionRetailAveragePrice','MMRCurrentAuctionAveragePrice', 'MMRCurrentRetailAveragePrice',
           'VehBCost','TopThreeAmericanName']]
data=data.dropna()
data.loc[data['Transmission']=='AUTO', 'Transmission'] =1
data.loc[data['Transmission']=='MANUAL', 'Transmission'] =0
data.loc[data['Transmission']=='Manual', 'Transmission'] =0
data.loc[data['TopThreeAmericanName']=='OTHER', 'TopThreeAmericanName'] =0
data.loc[data['TopThreeAmericanName']=='FORD', 'TopThreeAmericanName'] =1
data.loc[data['TopThreeAmericanName']=='CHRYSLER', 'TopThreeAmericanName'] =1
data.loc[data['TopThreeAmericanName']=='GM', 'TopThreeAmericanName'] =1

data['MMRAcquisitionAuctionAveragePrice'] = data['MMRAcquisitionAuctionAveragePrice']*100 /data['VehBCost']
data['MMRAcquisitionRetailAveragePrice'] = data['MMRAcquisitionRetailAveragePrice']*100 /data['VehBCost']
data['MMRCurrentAuctionAveragePrice'] = data['MMRCurrentAuctionAveragePrice'] *100/data['VehBCost']
data['MMRCurrentRetailAveragePrice'] = data['MMRCurrentRetailAveragePrice']*100 /data['VehBCost']
data['VehOdo'] = data['VehOdo'] /100000


target=data[['IsBadBuy']]
data=data.drop(['IsBadBuy','VehBCost'],axis=1)
x_train, x_test, y_train, y_test = data[:70000],data[70000:],target[:70000],target[70000:]
x_test
```

## Построение SVC

```python
%%time
from sklearn.svm import SVC
import warnings; warnings.simplefilter('ignore')

model = SVC(kernel='rbf')
model.fit(x_train, y_train)
pred=model.predict(x_test)
print(metrics.classification_report(y_test, pred))
```
